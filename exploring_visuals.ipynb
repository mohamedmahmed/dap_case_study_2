# Exploring with Visuals
Use `clean_08.csv` and `clean_18.csv`. You should've created these data files in the previous section: *Fixing Data Types Pt 3*.

```python
# load datasets
import pandas as pd

df_08 = pd.read_csv('clean_08.csv')
df_18 = pd.read_csv('clean_18.csv')

df_08.info()
```

```python
#  compare greenhouse gas score in 2008 and 2018
df_08['greenhouse_gas_score'].hist();
df_18['greenhouse_gas_score'].hist();
```

```python
df_08['cmb_mpg'].hist();
df_18['cmb_mpg'].hist();
```

```python
# correlation between displacement and combined mpg
df_08.plot.scatter(x='displ', y='cmb_mpg')
df_18.plot.scatter(x='displ', y='cmb_mpg')
```
